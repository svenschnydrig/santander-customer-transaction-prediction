## Light GBM

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split


## Importing data

df_train = pd.read_csv('/content/drive/MyDrive/Data Science Project - Team D/data/raw-data/train.csv')
df_test = pd.read_csv('/content/drive/MyDrive/Data Science Project - Team D/data/raw-data/test.csv')
submission = pd.read_csv('/content/drive/MyDrive/Data Science Project - Team D/data/raw-data/sample_submission.csv')

## Training the model

Y_train = df_train["target"]
X_train = df_train.iloc[:,2:]

x_train, x_test, y_train, y_test = train_test_split(X_train, Y_train, test_size=0.30)

lgb_train = lgb.Dataset(x_train, y_train)
lgb_eval = lgb.Dataset(x_test, y_test, reference=lgb_train)

params = {
    'boosting': 'gbdt',
    'objective': 'binary',
    'metric': 'auc'}

model = lgb.train(params, train_set=lgb_train,
                 valid_sets=lgb_eval,
                 early_stopping_rounds=20, verbose_eval= 10)



y_pred = model.predict(df_test.iloc[:,1:])

sub = pd.DataFrame(y_pred)
submission["target"] = sub
submission.to_csv("/content/drive/MyDrive/Data Science Project - Team D/submission/submission_basic_lgbm.csv", index=False)
```

## Logistic regression
